```python
import matplotlib.pyplot as plt
import string
import pandas as pd
import re
```

```python
hobbit_file_path = ('Resources/00 - Hobbit_or_There_and_Back_Again.txt')
fellowship_file_path = ('Resources/01 - The Fellowship Of The Ring.txt')
towers_file_path = ('Resources/02 - The Two Towers.txt')
return_file_path = ('Resources/03 - The Return Of The King.txt')

"""def read_book(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read().lower()"""
    
def read_book(file_path):
    # Try opening the file with utf-8 encoding
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            return file.read().lower()
    except UnicodeDecodeError:
        # If utf-8 fails, try another common encoding like ISO-8859-1
        with open(file_path, 'r', encoding='ISO-8859-1') as file:
            return file.read().lower()

def preprocess_text(text):
    translator = str.maketrans('','', string.punctuation)
    return text.translate(translator).split()

def get_word_length(words):
    return [len(word) for word in words if word.isalpha()]
```

### Word-Length Study of *The Hobbit*

```python
hobbit_text = read_book(hobbit_file_path)
hobbit_text
```

```python
hobbit_words = preprocess_text(hobbit_text)
hobbit_words
```

```python
hobbit_word_len = get_word_length(hobbit_words)
hobbit_word_len
```

```python
plt.hist(hobbit_word_len, bins=range(1, max(hobbit_word_len)+1))
plt.title("Word Length Frequency in 'The Hobbit'")
plt.xlabel('Word Length')
plt.ylabel('Frequency')
plt.show()
```

```python
def generate_histogram(word_length, title):
    plt.hist(word_length, bins=range(1, max(word_length)+1))
    plt.title(title)
    plt.xlabel('Word Length')
    plt.ylabel('Frequency')
    return plt

generate_histogram(hobbit_word_len, "Word Length Frequency in 'The Hobbit'")
```

### Comparison of Word-Length

```python
def avg_word_length(word_length):
    sum_of_lengths = sum(word_length)
    total_words = len(word_length)
    avg_word_length = sum_of_lengths / total_words
    return avg_word_length

print(f"Average Word-Length in 'The Hobbit' is {avg_word_length(hobbit_word_len):.2f}")
print(f"Average Word-Length in 'The Fellowship of the Ring' is {avg_word_length(fellowship_word_len):.2f}")
print(f"Average Word-Length in 'The Two Towers' is {avg_word_length(towers_word_len):.2f}")
print(f"Average Word-Length in 'The Return of the King' is {avg_word_length(return_word_len):.2f}")
```

The results of this word-length comparison are very surprising, because 'The Hobbit' is considered the book that is most-appropriate for children, and the story that younger readers are encouraged to begin with, as they journey for the first time into Middle Earth.
